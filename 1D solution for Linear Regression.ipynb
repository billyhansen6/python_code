```python
# load data and convert to numpy arrays
df = pd.read_csv('data_1d.csv', header = None)
df
```

```python
X = df[0]
```

```python
X = np.array(X)
```

```python
X.shape
```

```python
Y = df[1]
Y = np.array(Y)
```

```python
Y.shape
```

```python
# plot data.

plt.scatter(X,Y)
plt.show()
```

```python
# calculate a and b.

denominator = X.dot(X) - X.mean()*X.sum()
a = (X.dot(Y) - Y.mean()*X.sum()) / denominator
b = (Y.mean()* X.dot(X) - X.mean()* X.dot(Y)) / denominator
```

```python
# calculate predicted Y

Yhat = a*X + b
```

```python
plt.scatter(X,Y)
plt.plot(X, Yhat)
```

```python
# calculate r-squared
d1 = Y-Yhat
d2 = Y-Y.mean()
r2 = 1-d1.dot(d1)/d2.dot(d2)
print("the r-squared is:", r2)
```
